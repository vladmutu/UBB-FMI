```python
def samebirthday(n,nr_sims=5000,probability=False):
    if probability:
        p=1   # p Aceasta va reprezenta probabilitatea ca toate cele n persoane să aibă zile de naștere diferite.
        for k in range(n):
            p*=(365-k)/365  
# (365 - k) reprezintă numărul de zile disponibile pentru ca noua persoană să aibă o zi de naștere diferită de cele deja alese de persoanele anterioare.
# fracția (365 - k) / 365 este probabilitatea ca a k-a persoană să aibă o zi de naștere diferită de cele deja alese.
# De exemplu, pentru k = 0, fracția este 365/365 = 1, ceea ce reprezintă probabilitatea ca prima persoană să aibă orice zi de naștere.
# (365 - 1) / 365 = 364/365. Probabilitatea ca a doua persoană să aibă o zi de naștere diferită de prima este 364/365.
    return 1-p
# 1 - p reprezintă probabilitatea ca, într-un grup de n persoane, cel puțin două dintre ele să aibă aceeași zi de naștere.
```

```python
samebirthday(23,nr_sims=2000,probability=True)
```

```python
def patrat(nr_pcte=1000,cerinta=1):
    axis('square')
    axis((0, 1, 0, 1))
    count=0
    A,B,C,D,E=[0,0],[1,0],[1,1],[0,1],[0.5,0.5]
    match cerinta:
        case 1: 
            for _ in range(nr_pcte):
                F=[random(),random()]
                if dist(F,E)<0.5:
                    count=count+1
                    plot(F[0],F[1],'ro')
                else:
                    plot(F[0],F[1],'bo')
        case 2: 
            for _ in range(nr_pcte):
                F=[random(),random()]
                if dist(F,E)<min([dist(F,A),dist(F,B),dist(F,C),dist(F,D)]):
                    count=count+1
                    plot(F[0],F[1],'ro')
                else:
                    plot(F[0],F[1],'bo')
        case 3: 
            for _ in range(nr_pcte):
                F=[random(),random()]
                FA,FB,FC,FD=dist(F,A),dist(F,B),dist(F,C),dist(F,D)
                tri_obtuze=(FA**2+FB**2<1)+(FB**2+FC**2<1)+(FC**2+FD**2<1)+(FD**2+FA**2<1)
                if tri_obtuze==2:
                    count=count+1
                    plot(F[0],F[1],'ro')
                else:
                    plot(F[0],F[1],'bo')
    return count/nr_pcte
```
