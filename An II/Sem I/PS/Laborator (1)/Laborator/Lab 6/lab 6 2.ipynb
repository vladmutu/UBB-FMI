# Problema 1

```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot
```

```python
data = norm.rvs(loc=165,scale=10,size=5000)

hist(data, bins=14, density = True,range=(130,200),label='frecvențe relative')

x = linspace(130,200,1000)
plot(x,norm.pdf(x,loc=165,scale=10),'r-',label='funcția de densitate')

xticks(range(130,200,5))
legend(loc='upper right')
grid()
show()
```

```python
data.mean(), data.std()
```

```python
mean((160<=data)&(data<=170)), norm.cdf(170,loc=165,scale=10)-norm.cdf(160,loc=165,scale=10), sum((160<=data)&(data<=170))/5000
```

# Problema 2

```python
from scipy.stats import expon, uniform
from numpy import mean, std

n = 5000
r = uniform.rvs(size = n)
data = expon.rvs(loc=0,scale=5,size=n)*(r<0.4)+uniform.rvs(loc=4,scale=2,size=n)*(r>=0.4)     # alpha = 1/5 deci scale = 5
```

```python
data.mean(), data.std()
```

```python
mean(data<5),     0.4*expon.cdf(5,loc=0,scale=5)+0.6*uniform.cdf(5,loc=4,scale=2)
```

# Problema 3

```python
from scipy.stats import uniform
from numpy import exp, mean
from scipy.integrate import quad

n=5000
u = uniform.rvs(loc=-1,scale=4,size=n)   #  scale=4 (b-a)  adica (3-(-1))
g = lambda x : exp(-x**2)
mean(4*g(u))
```

```python
quad(g,-1,3)[0]
```
